```python
# import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the configuration of the plot we will create.

# read in the data using pandas library
# creating dataframe

df = pd.read_csv(r'D:\Google_Data_Analytics\8_Case_Studies\2_week\6_Alex_Data_Project\4_Using_Python\movies.csv')
```

```python
# let's look into the data by using head function
df.head()
```

```python
for col in df.columns:
    # np.mean - give the average 
    # isnull - check the null values in the columns in the dataframes
    pct_missing = np.mean(df[col].isnull())
    
    # .format - used to subsitute the variable in the print funcitons
    print('{} - {}%'.format(col, pct_missing))
```

```python
# .dtypes - used to view the datatypes of our column
df.dtypes # object - is string datatype in pandas
```

```python
# change the datatypes of columns using ".astype" 
# to remove the number with zero
# null values in the column so used 'fillna' - function fill zero then change the datatype

df['budget'] = df['budget'].fillna(0).astype(int)

df['gross'] = df['gross'].fillna(0).astype(int)
```

```python
df.head()
```

```python
# Create correct year column from the released column
# change the data type to string and slicing to get the year

# June 13, 1980 (United States) - first separated the date and country then separate the year from the date
# str[1] - is list because 'split' return as list

df['year_correct'] = df['released'].str.split('\(').str[0].str.split(',').str[1]

df.head()
```

```python
# sort the data frame by gross in descending
# ref link = 'https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html'
df.sort_values(by='gross', ascending=False)
```

```python
# set option funciton used to view the full data and by scroll to see them
# df.set_option('display.max_rows')

# drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=True)

# for dropping duplicate from the dataframe
df.drop_duplicates()
```

```python
# find the correlation in the dataset
# budget high correlation
# company high correlation
```

```python
# scatterplot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

# adding detail to the data
plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for flim')
# to visualize use show command
plt.show()
```

```python
# plot budget vs gross using seaborn
# link - 'https://seaborn.pydata.org/generated/seaborn.regplot.html'
sns.regplot(x='budget', y='gross', data=df, scatter_kws = {'color':'red'}, line_kws= {'color': 'blue'})
```

```python
# lets start looking at correlation
# link - 'https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html'
# it will work only for numeric values
df.corr(method='pearson') # different methods for correlation - pearson, kendall, spearman
```

```python
correlation_matrix = df.corr(method='pearson')

# visualize with heatmap
sns.heatmap(correlation_matrix, annot=True)
# adding detail to the data
plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# assign the unique value to the 'object that mean string datatype'
# because we need to company correlation

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category') # changing the data type
        # assigning the random code to the changed data type 
        df_numerized[col_name] = df_numerized[col_name].cat.codes
```

```python
df_numerized
```

```python
# every single field representation
numerized_matrix = df_numerized.corr(method='pearson')

# visualize with heatmap
sns.heatmap(numerized_matrix, annot=True)
# adding detail to the data
plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# to see correlation with every values
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# sort in order and paired

sorted_pairs = corr_pairs.sort_values()
```

```python
sorted_pairs
```

```python
# to find the high correlation
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
```
